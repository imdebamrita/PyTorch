```python
import torch
```

```python
x = torch.randn(3, requires_grad=True)
print(x)
```

```python
y = x+2
print(y)
```

```python
z = y*y*2
# z = z.mean()
print(z)
```

```python
v = torch.tensor([0.1, 1.0, 0.001], dtype=torch.float32)
z.backward(v) # dz/dx
print(x.grad)
```

```python
# x.requires_grad_(False)
# x.detach()
# with torch.no_grad()

x = torch.randn(3, requires_grad=True)
print(x)
print(x + 2)

x.requires_grad_(False)
print(x)

y = x.detach()
print(y)

with torch.no_grad():
  z = x + 2
  print(z)
```

```python
weights = torch.ones(4, requires_grad=True)

for epochs in range(3):
  model_output = (weights*3).sum()

  model_output.backward()

  print(weights.grad)

  weights.grad.zero_()
```
