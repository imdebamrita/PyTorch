```python
import torch
import numpy as np
```

```python
x  = torch.empty((4, 2, 3))
print(x)
```

```python
x = torch.ones(2,3)
print(x)
print(x.dtype)
```

```python
x = torch.ones(2,3, dtype=torch.int)
print(x)
print(x.dtype)
print(x.size())
```

```python
x = torch.tensor([2.5, 0.1])
print(x)
```

```python
x = torch.rand(2,2)
y = torch.rand(2,2)
print(x)
print(y)

print("#### ADD ####")

z = x + y
w = torch.add(x, y)
print(z)
print(w)

print("#### SUB ####")

z = x - y
w = torch.sub(x, y)
print(z)
print(w)

print("#### MUL ####")

z = x * y
w = torch.mul(x, y)
print(z)
print(w)

print("#### DIV ####")

z = x / y
w = torch.div(x, y)
print(z)
print(w)
```

```python
x = torch.rand(5, 3)
print(x)
print(x[:, 0])
print(x[1, :])
print(x[1, 1].item())
```

```python
x = torch.rand(4, 4)
print(x)
y = x.view(-1, 8)
print(y)
```

```python
a = torch.ones(5)
print(a)
b= a.numpy()
print(type(b))

a.add_(1)
print(a)
print(b)
```

```python
a = np.ones(5)
print(a)
b= torch.from_numpy(a)
print(b)

a += 1
print(a)
print(b)
```
